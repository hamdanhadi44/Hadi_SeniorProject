```python
import nltk
from nltk.tokenize import word_tokenize

# Download NLTK tokenizer data (if not already downloaded)
nltk.download('punkt')
x=0
def count_tokens(text):
    global x
    if x%100==0:
        print(x)
    x=x+1
    # Tokenize the text
    tokens = word_tokenize(text)
    # Return the number of tokens
    return len(tokens)

```

```python
#Counting how many paragraphs
def count_paragraphs(text):
    # Split the text based on double line breaks
    paragraphs = text.split('\n\n')

    # Remove empty paragraphs (e.g., consecutive line breaks)
    non_empty_paragraphs = [paragraph.strip() for paragraph in paragraphs if paragraph.strip()]

    # Count the number of paragraphs
    num_paragraphs = len(non_empty_paragraphs)

    return num_paragraphs
```

```python
def extract_count_from_counter(counter_str):
    """
    Extracts the integer count value from a string representation of a Counter object.
    
    Parameters:
        counter_str (str): String representation of the Counter object.
    
    Returns:
        int: Integer count value.
    """
    # Convert the string to a tuple using ast.literal_eval
    counter_tuple = ast.literal_eval(counter_str)
    
    # Extract the integer value
    count_value = counter_tuple[1]  # The count value is expected to be the second element of the tuple
    return count_value
```

```python
import spacy

# Load the spaCy English model
nlp = spacy.load("en_core_web_sm")
x=0
def count_named_entity_types(text):
    global x
    if x%100==0:
        print(x)
    x=x+1
    # Process the text with spaCy
    doc = nlp(text)
    
    # Initialize a dictionary to count named entity types
    entity_types_count = {}
    
    # Iterate over named entities in the document
    for ent in doc.ents:
        # Increment the count for the entity type
        entity_types_count[ent.label_] = entity_types_count.get(ent.label_, 0) + 1
    
    return entity_types_count
```

# Average Word Length

```python
def average_word_length(text):
    words = text.split()  # Split the text into words
    total_length = sum(len(word) for word in words)  # Calculate the total length of all words
    num_words = len(words)  # Count the number of words
    if num_words == 0:
        return 0  # To avoid division by zero if there are no words
    return total_length / num_words
```

```python
data['avg_word_length']=data['text'].apply(lambda x:average_word_length(x))
```

# Dictionary for count of every type of POS tagging

```python
import nltk
from collections import defaultdict
x=0
def pos_tag_counts(text):
    global x
    if x%100==0:
        print(x)
    x=x+1
    # Tokenize the text into words
    words = nltk.word_tokenize(text)
    
    # Perform POS tagging
    pos_tags = nltk.pos_tag(words)
    
    # Initialize a defaultdict to store counts
    tag_counts = defaultdict(int)
    
    # Count the occurrences of each POS tag
    for _, pos in pos_tags:
        tag_counts[pos] += 1
    if x%100==0:
        print(tag_counts)
    return dict(tag_counts)


```

```python
data['pos_tag_dict']=data['text'].apply(lambda x: pos_tag_counts(x))
```

```python
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
x=0
def clean_text(text):
    global x
    if x%100==0:
        print(x)
    x=x+1
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Convert text to lowercase
    text = text.lower()
    
    # Tokenize text
    tokens = word_tokenize(text)
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    
    # Lemmatize words
    lemmatizer = WordNetLemmatizer()
    cleaned_text = [lemmatizer.lemmatize(word) for word in filtered_tokens]
    
    # Join tokens back into a single string
    cleaned_text = ' '.join(cleaned_text)
    
    return cleaned_text
```

```python
data['clean_text']=data['text'].apply(lambda x: clean_text(x))
```

# Topic Modeling Clustering

```python
'''Topic modeling clustering, also known as topic-based clustering, is a technique that 
combines topic modeling and clustering methods to group similar documents based on their thematic content. '''
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import Normalizer
from gensim.parsing.preprocessing import preprocess_string
from gensim.corpora import Dictionary

# Preprocess text
def preprocess_text(text):
    return preprocess_string(text)
print(1)
data['preprocessed_text'] = data['text'].apply(preprocess_text)
print(2)

# Create a dictionary and corpus
dictionary = Dictionary(data['preprocessed_text'])
corpus = [dictionary.doc2bow(text) for text in data['preprocessed_text']]
print(3)

# Convert the preprocessed text back to strings
preprocessed_texts = [' '.join(text) for text in data['preprocessed_text']]
print(4)

# Apply LDA
lda_model = LatentDirichletAllocation(n_components=5, random_state=42)
topic_vectors = lda_model.fit_transform(CountVectorizer().fit_transform(preprocessed_texts))
print(5)

# Cluster documents based on topic distributions
#5 clusters is the optimal number of clusters based on the elbow method
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(topic_vectors)

# Assign cluster labels to DataFrame
data['topic_cluster'] = clusters

```

# Standard Deviation of Sentences length for a text

```python
import nltk
import numpy as np

# Ensure you have NLTK downloaded with the required corpora
# nltk.download('punkt')

# Function to calculate standard deviation of sentence lengths
def sentence_length_std(text):
    # Tokenize the text into sentences
    sentences = nltk.sent_tokenize(text)
    
    # Compute the length of each sentence
    sentence_lengths = [len(nltk.word_tokenize(sentence)) for sentence in sentences]
    
    # Calculate the standard deviation of sentence lengths
    std_dev = np.std(sentence_lengths)
    
    return std_dev

data['sentence_lengths_std']=data['text'].apply(lambda x: sentence_length_std(x))
```
