```python
ai=data[data['generated']==1]
human=data[data['generated']==0]
```

# Stop Ratio

```python
#Analyzing the stop ratio for Ai and human texts

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd



# Calculate the mean, std, and median 'stop_ratio' for each group
grouped_stats = data.groupby('generated')['stop_ratio'].agg(['mean', 'std', 'median']).reset_index()

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

# Plot mean
sns.barplot(x='generated', y='mean', data=grouped_stats, color="skyblue", ax=axes[0])
axes[0].set_title("Mean stop_ratio")
axes[0].set_xlabel("Generated Group")
axes[0].set_ylabel("Mean stop_ratio")

# Plot std
sns.barplot(x='generated', y='std', data=grouped_stats, color="lightgreen", ax=axes[1])
axes[1].set_title("Standard Deviation stop_ratio")
axes[1].set_xlabel("Generated Group")
axes[1].set_ylabel("Standard Deviation stop_ratio")

# Plot median
sns.barplot(x='generated', y='median', data=grouped_stats, color="salmon", ax=axes[2])
axes[2].set_title("Median stop_ratio")
axes[2].set_xlabel("Generated Group")
axes[2].set_ylabel("Median stop_ratio")

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python

sns.boxplot(data=data, x='generated',y='stop_ratio')
```

```python

# Define a function to calculate mean without outliers
def mean_without_outliers(column):
    q1 = column.quantile(0.25)  # First quartile
    q3 = column.quantile(0.75)  # Third quartile
    iqr = q3 - q1  # Interquartile range
    lower_bound = q1 - 1.5 * iqr  # Lower bound for outliers
    upper_bound = q3 + 1.5 * iqr  # Upper bound for outliers
    filtered_column = column[(column >= lower_bound) & (column <= upper_bound)]  # Filter outliers
    return filtered_column.mean()


print('ai ' ,mean_without_outliers(ai['entity_ratio']))
print('human ',mean_without_outliers(human['entity_ratio']))


```
